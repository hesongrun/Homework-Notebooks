```python
tree = ET.parse('../../../JPE/JPE.xml')
```

```python
root = tree.getroot()
```

```python
# for tig in root.findall('./rec/header/controlInfo/artinfo/tig/atl'):
#     print(tig.tag, tig.text)
```

```python
article_list = []
cls_list_agg = []
for i in root.findall('./rec/header/controlInfo/artinfo'):
    article = i.findall('tig/atl')
    article_list.append(article[0].text)
#     print(article[0].text)
#     print(' ')
    try:
        cls_list = [dict_lv3[j.text.upper()] for j in i.findall('su')]
        cls_list_agg.append(cls_list)
    except KeyError:
        cls_list = []
        cls_list_agg.append(cls_list)
        continue
```

```python
cls_list_agg
```

```python
df_JPE = pd.DataFrame({'title': article_list, 'JEL': cls_list_agg})
```

```python
df_JPE['JEL_len'] = df_JPE['JEL'].apply(len)
```

```python
df_JPE = df_JPE[df_JPE['JEL_len']!=0]
```

```python
df_JPE.drop(columns = 'JEL_len', inplace=True)
```

```python
df_JPE
```

```python
df_text = pd.read_csv('../../../../TTIC31220/wos_econ_text_filtered.csv')
```

```python
df_text = df_text[df_text['source']=='JOURNAL OF POLITICAL ECONOMY']
```

```python
df_JPE['title'] = df_JPE['title'].str.upper()
df_text['title'] = df_text['title'].str.upper()
```

```python
df_JPE_JEL = pd.merge(df_text[['title', 'abstract']], df_JPE, on = 'title')
```

```python
def get_first(x):
    return x[0][0]
df_JPE_JEL['JEL_first'] = df_JPE_JEL['JEL'].apply(get_first)
```

```python
dict_lv1_inv = {}
for i in range(len(df_lv1)):
    dict_lv1_inv[df_lv1.iloc[i, 0]] = df_lv1.iloc[i, 1].upper()
```

```python
def get_cls_text(x):
    return dict_lv1_inv[x]
```

```python
df_JPE_JEL['Class'] = df_JPE_JEL['JEL_first'].apply(get_cls_text)
```

```python
df_JPE_JEL.to_pickle('JPE_JEL.pkl')
```
